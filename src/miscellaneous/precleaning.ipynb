```python
import pandas as pd
import numpy as np

```

```python
df= pd.read_csv('footballer_data.csv')
```

```python
df.columns
```

```python
# Reformat `Weight` to include " kg"
df['Weight'] = df['Weight'].astype(str) + " kg"

# Reformat `Height` to include " cm"
df['Height'] = df['Height'].astype(str) + " cm"

# Reintroduce special formatting for `Value`
def format_currency(value):
    if value >= 1_000_000:
        return f"€{value / 1_000_000:.1f}M"
    elif value >= 1_000:
        return f"€{value / 1_000:.1f}K"
    else:
        return f"€{value}"
```

```python
df['Value'] = df['Value'].apply(format_currency)

# Reintroduce special formatting for `Wage`
```

```python
df['Wage'] = df['Wage'].apply(format_currency)

```

```python
# Reintroduce special formatting for `Release Clause`
df['Release clause'] = df['Release clause'].apply(format_currency)

# Combine `Team` and `Contract` back into `Team & Contract`
df['Team & Contract'] = df['team'] + "\n" + df['contract']
```

```python
df.to_csv("unclean_dataset.csv", index=False)
```

```python
import re
```

```python
df['Weight'] = df['Weight'].str.split(' ').str[0]
pattern = re.compile(r'\D+')
df['Weight'] = [int(pattern.sub('', weight)) for weight in df['Weight']]
```

```python

df['Height'] = df['Height'].str.split(' ').str[0]
df['Height'] = [int(pattern.sub('', height)) for height in df['Height']]
```

```python

df['new_values'] = df["Value"].str[1:]
# Her bir değerin sayısal değerini hesaplama
new_values = []
for value in df['new_values']:
    if 'M' in value:
        new_values.append(float(value.replace('M', '')) * 1000000)
    elif 'K' in value:
        new_values.append(float(value.replace('K', '')) * 1000)
    else:
        new_values.append(float(value))

# Sonuçları yazdırma
df['new_values_series'] = new_values
df.head()
```

```python
for index, row in df.iterrows():
    team_contract = row["Team & Contract"]
    team = team_contract.split("\n")[0]
    df.at[index, "Team"] = team
df.head()
```

```python

df['new_wages'] = df["Wage"].str[1:]
# Her bir değerin sayısal değerini hesaplama
new_wages = []
for value in df['new_wages']:
    if 'M' in value:
        new_wages.append(float(value.replace('M', '')) * 1000000)
    elif 'K' in value:
        new_wages.append(float(value.replace('K', '')) * 1000)
    else:
        new_wages.append(float(value))

# Sonuçları yazdırma
df['new_wages_series'] = new_wages
df.head()
```
